```python
import pandas as pd
import datashader as ds
import datashader.transfer_functions as tf
import bokeh.plotting as bp
from datashader.bokeh_ext import InteractiveImage
import numpy as np
from bokeh.io import output_notebook, show
from bokeh.plotting import figure

output_notebook()
```

```python
#df = pd.read_csv("https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/ggplot2/diamonds.csv", index_col=0)
df = pd.read_csv("diamonds.csv", index_col=0)
df.head(5)
```

```python
p = figure(width=500, height=500)
cuts = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
colors = ["red", "blue", "green", "brown", "yellow"]
for cut, color in zip(cuts, colors):
    data = df[df["cut"] == cut]
    p.circle(data["x"], data["carat"], size=2, color=color, alpha=0.2)
show(p)
```
