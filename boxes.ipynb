```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df_male = pd.read_csv('male-names.csv', sep=';')
df_female = pd.read_csv('female-names.csv', sep=';')
df = pd.concat([df_male, df_female])
display(df.describe())
df.head(5)
```

```python
data = df[df["Name"].isin(["Виктор", "Татьяна", "Александр"])]
sns.boxplot(x="Name", y="NumberOfPersons", data=data, whis=1.5)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
display(data.describe())
plt.show()
```

```python
def box_name(name):
    d = df[df["Name"] == name]
    sns.boxplot(x="Name", y="NumberOfPersons", data=d, whis=1.5)
    display(d.describe())
    plt.show()
```

```python
box_name("Виктор")
```

```python
box_name("Александр")
```

```python
box_name("Татьяна")
```

```python
dt = df[df["Name"] == "Татьяна"]
stats = dt.describe()["NumberOfPersons"]
Q1, Q3 = stats["25%"], stats["75%"]
IQR = Q3 - Q1
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print()
print("Умеренные выбросы")
display(dt[(dt["NumberOfPersons"] > Q3 + 1.5 * IQR) & (dt["NumberOfPersons"] <= Q3 + 3 * IQR)])
print("Экстремальные выбросы")
display(dt[dt["NumberOfPersons"] > Q3 + 3 * IQR])
```
