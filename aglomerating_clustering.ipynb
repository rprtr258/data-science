```
#@title Silhoutte score for agglomerative clustering
from sklearn.metrics import silhouette_score

ns = np.arange(2, 10)
ss = []

for n in ns:
    clustering = AgglomerativeClustering(n_clusters=n)
    clustering.fit(X)
    ss.append(silhouette_score(X, clustering.labels_))
plt.plot(ns, ss)
plt.show()
```
